# *Kenar Tespit Algoritması*

En temel anlamıyla görüntü işlemede bir kenar, piksel yoğunluk değerlerinde ani bir değişimin meydana geldiği bir dizi bitişik pikseller bütünüdür. Kenarlar ön plan ve arka plan resimler arasındaki sınırları temsil eder ve görüntü işleme, makine öğrenmesi, bilgisayar görmesinde bir çok işlemin altyapısını oluşturmaktadır. Kenar algoritmaların temelde altında yatan işlemler, görüntü üzerinde meydana gelen ani piksel değeri değişimleri veya ani parlaklık, yoğunluk değişimlerini yakalayacak bir dizi matematiksel formüllerdir. Gerçek bir resimde kenar olmadığı halde kenar olarak algılanmasına neden olan gölgeler ve benzeri unsurlar veya kenar olduğu halde kenar olarak algılanmasını engelleyecek parlaklık ve benzeri unsurlar bulunabilmektedir.

Kenar tespit algoritmaları genelde 2 adımda çalışır.
Adım: Kenar bir piksellik bir değişim olduğu için resimde bulunan farklı gürültülerin kenar olarak algılaması muhtemeldir. Engel olmak için kenar tespit algoritmanın da ilk adım filtrelemedir. Filtrelemeden kasteden bir blurring işlemi olaiblir. Gereğinden fazla yapılacak bir filtreleme işleminde kenar çizgileri kaybolabilir veya gereğinden az yapılan bir filtreleme ise herhangi bir sonuç değişimi yaşanmayacağından sadece repormans kaybına neden olabilir.

Anlamlı yoğunluk değişimlerinin daha iyi yakalanması için kenar belirginleştirme sürecidir. Bu adımda resmi uygulanacak bir kernel ile kenar noktalarının olduğu alanlar belirginleştirir. Ancak kenar olmayan veya yoğunluk değişiminin anlamlı olmadığı, alanların kaldırılması için eşitleme kullanılmaktadır.

Kenar tespit algoritmaların genel olarak 3 grupta inceleyebiliriz. Birinci grup algoritmalar yoğunluk değişim tespiti için görüntünün birinci türevini kullanır.(Sobel, Perwit, Roberts)

İkinci Grup algoritmalar 2. türevi kullanan Laplacian ve second directional derivative operatörlerdir. Görüntülere önce filtrenin işlevini yapar ve farklı parametrelerle ayarlarını değiştirmesine imkan sağlayan Canny ve Deriche algoritmaları ise 3.gruptur

**Sobel Operatörü**: En basit tanımıyla çalışması, gri renk kdolamasını çevrilmiş bir görüntünün her 3x3 piksellik alanında yatayda(4.1) ve dikeyde(4.2) tam sayı değerli kernelleri uygulamak ve sonrasında bunları birleştirmektir(4.3). Bazı durumlarda kenarlığın eğim yönü ve açısında bilinmesi gerekir(4.4)

Denklemleri buraya yaz.

Not: Sobel operatörü kenar yapısını belirgin hale getirirken, çoğu zaman kenar olmadığı halde kenar gibi hesaplanan ufak alanların ortaya çıkmasına sebep olabilmektedir. Bu Durumu engellemek için kenar çalışması yapmadan önce





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme
#Görüntüyü oku(siyah-beyaz)
image= cv2.imread("Data/19_table.png", cv2.IMREAD_GRAYSCALE)
#Sobel Operatörünü uygula
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
#ksize: Sobel kernelinin(filtresinin) büyüklüğünü ayarlamak için 1,3,5,7 rakamlarından birini alır
#(dx=1, dy=0, ksize=3), burada dx ve dy yatay ve dikey eksende kullanılacak dereceyi belirler
#Sobel operatörünün mutlak değerini al
sobelx= np.abs(sobelx)
sobely= np.abs(sobely)

#Dikey ve yatay kenarları birleştirir
edges= cv2.bitwise_or(sobelx, sobely)

#Kenarları görselleştir
cv2_imshow(image) #orijinal image
cv2_imshow(edges) #kenarlar
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme
#Görüntüyü oku(siyah-beyaz)
image= cv2.imread("Data/19_table.png", cv2.IMREAD_GRAYSCALE)

#Laplacian operatörü ile kenar tespiti
sonuc1 = cv2.Laplacian(image, cv2.CV_64F)

#Laplacian görüntüyü normalize et
sonuc1 = np.uint8(np.absolute(sonuc1))

imgBlured=cv2.GaussianBlur(image,(3,3),0)
sonuc2 = cv2.Laplacian(imgBlured, ddepth=-1, ksize=3)
#ddepth=-1 giriş resmi ile çıkış resmi aynı derinlikte, 3x3 kernel

f,eksen=plt.subplots(1,3,figsize= (17,5))
eksen[0].imshow(image, cmap="gray")
eksen[1].imshow(sonuc1, cmap="gray")
eksen[2].imshow(sonuc2, cmap="gray")
```

### **Canny Kenar Algoritması**:

Günümüz kameraları odaklama, netleştirme ve benzeri birçok özelliğe sahiptir. Bu özelliklerin farklı şekillerde ayıklanmasıyla görüntüler üzerindeki gürültülerin bastırılması(ancak kenarların kaybolması) veya benzer şekilde kenarların netleştirilmesi(ancak resmin gürültülü kalmasıyla) sonuçlanabilmektedir. Canny algoritması bu dengeyi iyi şekilde sağlamak için geliştirilmiştir. En basit tanımıyla, Gaussian filtresinin birinci türevi gibi düşünülebilir. Algoritma 5 farklı parça olarak çalışır

Gürültü Azaltma: Kenar Tespitinin gürültülere karşı oldukça hassas olduğundan bahsedilmişti. Uygulanacak yerler, büyüklüğü tek rakamlardan oluşmalıdır.

Gradyan Hesaplaması: Bu adımdaki amaç filtrelenmiş resimde piksellere meydana gelen ani yoğunluk değişimlerini tespit etmektir. Buradan sonuçla kenar pikselleri ve kenar açısı bulunur. Bu işlemlerde Sobel operatörü tercih edilir.

Maksimum Olmayan Bastırma: İkinci adım ile elde edilen kenarlarını inceltme işlemi yapılmalıdır. Bunun için bulunan bütün kenarlar incelenerek birleşik kenar piksellerinden yoğun olanlar ve olmayanlar tespit edilir. Bu işlemden sonra elde edilen piksel kalınlığı artık kenar çizgisi haline gelecektir.

Double treshold: Bir önceki adım ile elde edilen görünü üzerinde kimi kenarlar koyup kimileri açık renk kalırken, bazı kısımlarda ise gürültüden kaynaklı ince kenarların olduğu görülecektir. Bu adım ile amaçlanan kenar yapılarının bir bütünü kazanması ve kenar olarak algılanan istenmeyen piksellerin ortadan kaldırılması için piksellerin gruplandırılmasıdır. Pikseller 3 seviyeye ayrılır, ilk seviye pikseller kenar oldukları kesin olan pikseller üçüncü seviye pikseller kenar olmamsı gereken gürültüden kaynaklı tespit edilmiş pikseller, ikisinci seviye pikseller ise zayıf ancak içerisinde kenar olan ve olmayan pikselleri birlikte barındıran piksellerdir.

Histeresiz ile kenarların Tespiti: İkinci seviye pikselleden kenar olma ihtimali olanlar kenar olarak algılanır

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme
#Görüntüyü oku(siyah-beyaz)

image= cv2.imread("Data/21_bird.png", cv2.IMREAD_GRAYSCALE)
low_threshold= 50
high_threshold = 150
sonuc1= cv2.Canny(image, low_threshold, high_threshold, L2gradient= True)
#L2gradient=True gradyan hesabında yatay/dikey kenarların kareleri toplamının karekökü kullanılır.
f,eksen=plt.subplots(1,2,figsize=(17,7))
eksen[0].imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB), cmap="gray")
eksen[1].imshow(sonuc1, cmap="gray")
```

### **Deriche filtresi ile kenar tespiti**
OpenCV'de Deriche kenar tespiti algoritması için özel bir fonksiyon bulunmamaktadır. Ancak, deriche filtresini uygulamak için OpenCV'nin cv2.getDerivKernels fonksiyonunu kullanabilirsiniz. Özellikle gürültülü resimlerde çok daha iyi sonuçlar alınabilir.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme
#Görüntüyü oku(siyah-beyaz)

image= cv2.imread("Data/21_bird.png", cv2.IMREAD_GRAYSCALE)

#Deriche filteresi için kernel oluştur
alpha = 0.5 #Deriche filtresi parametresi
kernel_size = 3 # Filtre boyutu

kx, ky = cv2.getDerivKernels(1, 1, kernel_size, normalize=True)
deriche_kernel_x = alpha * kx
deriche_kernel_y= alpha * ky

#Görüntüyü deriche filtresi ile türevle
deriche_x = cv2.filter2D(image, -1, deriche_kernel_x)
deriche_y = cv2.filter2D(image, -1, deriche_kernel_y)

#Kenarları birleştir
edges = np.sqrt(deriche_x**2 + deriche_y**2)

f,eksen= plt.subplots(1,2, figsize = (17,7))
eksen[0].imshow(image, cmap="gray")
eksen[1].imshow(edges, cmap="gray")
```

### **Köşe Tespit Algoritmaları:**
Kenar tespitinde yatay ve dikede yan yana veya üst üste yoğunluk değişimlerinin olduğu pikseller aranırken, burada ise benzer şekilde ancak tüm yönlerde yoğunluk değişiminin olduğu pikseller aranmaktadır. Daha teknik bir tanımla, her bir pikselin komşu olduğu pikseller incelenecek ve aralarında anlamlı bir değişiklik varsa bu piksel için köşe tespiti yapılacaktır.

*Harris Köşe Tespit Algoritması:*
Resimde bulunan mavi kutucuk butün resim boyunca dolaştırlır ve tarama yapar. Tarama sonucunda tüm yönlerde piksel yoğunluğu değişimi tespit edildiği noktalar kenar pikseller olarak işaretlenir. Mavi kutucu a resminde olduğu gibi boşlukta olduğunu farz edelim. Kutucuk içeriği analiz edildiğinde içerisinde hiçbir yoğunluk değişiminin olmadığı görülecektir. B resminde ise yoğunluk değişimi sadece yatay yönde yer alacağından dolayı bu pikselde kenar piksel olarak işaretlenmeyecektir. Ancak c resminde ani yoğunluk değişimi bütün yönlerde olmaktadır. Bu sebeple burası bir kenar noktası tespiti için işaretlenecektir.

Bu algoritma renkli resimler üzerinde çalıştırılmaz. Resmin yatay ve dikey eksenlerde birinci dereceden türevli gradyan hesaplaması yapılır(Sobel Operatörü). Köşe değerlerinin hesaplanmasında denklem 4.10 kullanılmakadır. Denklemde m tarama yapılan alandaki kovaryans matrisidir. Eğer r değeri çok küçükse resim 4.6 olduğu gibi piksel yoğunluğu değişmeyen düz bir alan vardır. R değeri sıfırdan küçükse 4.6b'de olduğu gibi bir kenar tespit edilmiştir, köşe pikseli değildir. Ancak r değeri çok büyükse her yönde ani yoğunluk değişimi mevcuttur ve üzerinde çalışılan(x,y) pikseli bir köşe pikseldir.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme
#Görüntüyü oku(siyah-beyaz)

def harris_corner_detection(image_path):
  #Resmi yükle
  img = cv2.imread(image_path)
  gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  #Harris köşe tespiti için parametreleri ayarla
  corner_quality = 0.04
  min_distance=10
  block_size=3
  #Harris köşe tespiti uygula
  corners = cv2.cornerHarris(gray, block_size, 3, corner_quality)

  #Köşeleri belirli bir eşik değerine göre seç
  corners = cv2.dilate(corners,None)
  img[corners > 0.01 * corners.max()] = [0,0,255]

  #Sonucu göster
  cv2_imshow(img)

if __name__ == "__main__":
  image_path = "DATA/22_real_chess.jpg" # Kendi resminizin yolunu buraya ekleyin
  harris_corner_detection(image_path)
```

# **Kontur Tespit Algoritması(Contour Detection)**
Kontur, bir resim üzerinde bulunnan aynı renk ve yoğunluğa sahip kenar piksellerinin bir eğri olarak birleştirilmewsidir. Eğer bir nesnenin sınırlarını çıkarabilirsek nesnelerinin sayımını da gerçekleştirebiliriz. Gerçek resimlerle çalışırken blurring yapmak gereklidir. Yumuşatma işleminde kullanılacak kernel boyutunun resmin boyutuna göre arttırılması gerekir. Kernel çok büyük seçilirse de kenarların kaldırılması söz konusu olabilir. Yumuşatma işleminden sonraki treshold ise sınırların ortaya çıkması için ve gürültünün azaltılması için önemlidir. findContours() fonksiyonu için dri formatta resme ihtiyaç vardır. Bu fonksiyon bir resimdeki her bir nesne için sınır piksellerinin koordinatlarını ndarray nesnesi olarak geri döndürür.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme

def contour_detection(image_path):
  img=cv2.imread(image_path)
  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  edges = cv2.Canny(gray,50,150)
  contours,_=cv2.fintContours(edges,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  cv2.drawContours(img, contours, -1, (0,255,0), 2)

  cv2_imshow(img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

if __name__ == "__main__":
  image_path = "DATA/25_contour.jpg"
  contour_detection(image_path)
```

# **Watershed Algoritması**

Watershed Algoritması, Görüntü işlemede nesne segmentasyonu için kullanılan bir algoritmadır. Bu algoritma, bir görüntüdeki farklı nesneleri birbirinden ayırmak veya nesne sınırlarını belirmelek için kullanılır. Watershed Algoritması, görüntüyü bir topografik haritaya benzetir ve bu haritada havzaları bulmaya dayanır.

Segmentasyon işleminin nasıl yapıldığını anlamak için, haritayı kapalı bir kap olarak düşünelim. Mavi çizgilerle işaretlenmiş alanlardan sular yükselmeye başlayacaktır. Kırmızı noktalar tepe noktalarıdır. Haritada en alt(vadi) ve en üst(tepe)noktalarıdır. Sular yükseldikçe watershed adı verilen sınır bölgeleri yükselecek ve suların karışması engellenecektir. Bu sayede seviyelere göre görüntü üzerinde bulunan her bir segment tespit edilmiş ve birbirinden ayrılmış olacaktır. Benzer şekilde resimler üzerinde bulunnan bütün nesnelerin ayrılması sağlanmaktadır. Resim gri formata çevrilir. Yüüksek yoğunluklu bölgeler tepe, düşük yoğunluklu bölgeler vadi gibi düşünülerek parçalanabilir. Topolojideki tepe ve vadi görüntüde ön plan ve arka plan nesnelerine denk gelir. Sınır noktalarını belirlenmesi kritik bir noktadır. Bunun için 2 yaklaşım vardır. İlkinde gri formatta düşük yoğunlüklar ön plan resmi, yüksek yoğunlukta arka plan olarak etkilenir. Aradaki sınırlar ise algoritma tarafından hesaplanır. 2.yaklaşım ise ön plan ve arka plan kısımları manuel olarak işaretlerin. Burada ön plan ve arka plan olarak emin olduğumuz, emin olmadığımız noktaların işaretlenerek bir matriste tutulması gerekir




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab\Notebooks/Ders:GoruntuIsleme

imgOrj = cv2.imread('DATA/26_coins.jpg')
imgBlr = cv2.medianBlur(imgOrj,31)
imgGray= cv2.cvtColor(imgBlr, cv2.COLOR_BGR2GRAY)
ret, imgTH=cv2.threshold(imgGray,0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

kernel=np.ones((5,5), np.uint8)
imgOPN= cv2.morphologyEx(imgTH, cv2.MORPH_OPEN,kernel,iterations=7)
sureBG=cv2.dilate(imgOPN, kernel, iterations=3)

#ÖnPlan Alanı
#distanceTransform() ile her pikselin en yakın sıfır değerine sahip piksele olan mesafesi hesaplanır
#Nesnelerin merkez pikselleri yani sıfır piksellerine en uzak nokta beyaz kalırken, siyah piksellere yaklaştıkça piksel değerleri düşer.
#Böylece madeni para, yani emin olduğum ön plan piksellerin ortaya çıkması sağlanmıştır
dist_transform=cv2.distanceTransform(imgOPN,cv2.DIST_L2_5)

#Eşikleme yap
#Eşikleme değeri olarak hesaplanan maksimum mesafenin %70den büyük olanlarının piksel değeri 255 yapılarak sureFG elde edilmiştir

#Bilinmeyen Bölgeleri Bul
#Emin olduğumuz arkaplan ve ön plan arasında kalan alandır
sureFG= np.uint8(sureFG)
unknown = cv2.subtract(sureBG, sureFG)

#Etiketleme işlemi
ret, markers = cv2.connectedComponents(sureFG, labels=5)

#Bilinmeyen pikselleri etiketle
markers = markers + 1
markers[unknown==255] = 0

#Watershed algoritması uygula
markers = cv2. watershed(imgOrj, markers)

contours, hierarchy=cv2.findContours(markers, cv2.RETR_CComp, cv2.CHAIN_APPROX_SIMPLE)

imgCopy = imgOrj.copy()
for i in range(len(contours)):
  if hierarchy[0][i][3] ==-1:
    cv2.drawContours(imgCopy, contours, i, (255,0,0),5)

f,eksen = plt.subplots(3,3,figsize=(30,30))
eksen[0,0].imshow(imgOrj)
eksen[0,1].imshow(imgBlr)
eksen[0,2].imshow(imgGray)
eksen[1,0].imshow(imgTH)
eksen[1,1].imshow(imgOPN)
eksen[1,2].imshow(sureBG)
eksen[2,0].imshow(dist_transform)
eksen[2,1].imshow(sureFG)
eksen[2,2].imshow(imgCopy)
```
